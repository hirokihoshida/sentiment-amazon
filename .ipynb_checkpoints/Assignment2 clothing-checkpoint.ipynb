```python
import pandas as pd
import numpy as np
import math


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import RidgeClassifier
```

```python
dataO = pd.read_json('renttherunway_final_data.json', lines=True)
dataO
```

```python
def heightCleaner(height):
    if pd.isnull(height):
        return None
    h = height.split()
    h[0] = int(h[0].replace("'", ''))*12
    h[1] = int(h[1].replace('"', ''))
    return sum(h)
def weightCleaner(weight):
    if pd.isnull(weight):
        return None
    return weight.split('lbs')[0]

#logistic regression
data = dataO.copy()

data = data.drop(columns=['review_text', 'review_summary', 'review_date', 'bust size'])

data['height'] = data['height'].apply(heightCleaner)

data['weight'] = data['weight'].apply(weightCleaner)

data['rating'] = data['rating'] / 2

fit = pd.get_dummies(data['fit'], prefix='f')
data = pd.concat([data,fit], axis=1).drop(columns=['fit'])

category = pd.get_dummies(data['category'], prefix='c')
data = pd.concat([data,category], axis=1).drop(columns=['category'])

rentFor = pd.get_dummies(data['rented for'], prefix='r')
data = pd.concat([data,rentFor], axis=1).drop(columns=['rented for'])

bodyType = pd.get_dummies(data['body type'], prefix='b')
data = pd.concat([data,bodyType], axis=1).drop(columns=['body type'])


data
```

```python
data = data.dropna()
```

```python
X = data.drop('rating', axis=1)
y= data['rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
model = RidgeClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
pd.Series(model.predict(X_test)).unique()
```
